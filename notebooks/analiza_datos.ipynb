```python
def analisis_bivariante(df, col1, col2):
    """
    Realiza análisis bivariante entre dos columnas del DataFrame,
    seleccionando visualización adecuada según el tipo de datos.

    Parámetros:
    - df: pandas.DataFrame
    - col1: str, nombre de la primera columna
    - col2: str, nombre de la segunda columna
    """
    if col1 not in df.columns or col2 not in df.columns:
        raise ValueError("Una o ambas columnas no existen en el DataFrame.")

    tipo1 = df[col1].dtype
    tipo2 = df[col2].dtype

    es_num1 = pd.api.types.is_numeric_dtype(tipo1)
    es_num2 = pd.api.types.is_numeric_dtype(tipo2)
    es_cat1 = pd.api.types.is_categorical_dtype(tipo1) or pd.api.types.is_object_dtype(tipo1)
    es_cat2 = pd.api.types.is_categorical_dtype(tipo2) or pd.api.types.is_object_dtype(tipo2)

    plt.figure(figsize=(10, 6))

    # Numérica vs Numérica
    if es_num1 and es_num2:
        sns.scatterplot(data=df, x=col1, y=col2, alpha=0.7)
        plt.title(f'Scatter plot: {col1} vs {col2}')
        plt.xlabel(col1)
        plt.ylabel(col2)

        # Mostrar correlación
        correlacion = df[[col1, col2]].corr().iloc[0, 1]
        print(f"Correlación entre {col1} y {col2}: {correlacion:.2f}")

    # Categórica vs Numérica
    elif (es_cat1 and es_num2):
        sns.boxplot(data=df, x=col1, y=col2, palette="Set2")
        plt.title(f'Boxplot: {col2} por {col1}')
        plt.xlabel(col1)
        plt.ylabel(col2)

    elif (es_num1 and es_cat2):
        sns.boxplot(data=df, x=col2, y=col1, palette="Set2")
        plt.title(f'Boxplot: {col1} por {col2}')
        plt.xlabel(col2)
        plt.ylabel(col1)

    # Categórica vs Categórica
    elif es_cat1 and es_cat2:
        tabla = pd.crosstab(df[col1], df[col2])
        sns.heatmap(tabla, annot=True, fmt='d', cmap='YlGnBu')
        plt.title(f'Tabla de contingencia: {col1} vs {col2}')
        plt.xlabel(col2)
        plt.ylabel(col1)

    else:
        print("Combinación de tipos no soportada para análisis bivariante.")
        return

    plt.tight_layout()
    plt.show()
```
